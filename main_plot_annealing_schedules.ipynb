# Plotting Annealing Schedules

## T_start: 2.0, T_freeze: 0.15, N_steps: 120000

### Multiplicative Schedules  

```python
t_start = 2.0
t_freeze = 0.15
N_steps = 120000

schedule = ConstantSchedule(2.0)
plot_schedule( schedule.get_schedule(N_steps), 'Constant Schedule' )

schedule = ExponentialMultiplicativeSchedule( T_start=t_start, T_freeze=t_freeze , alpha=0.80)
plot_schedule( schedule.get_schedule(N_steps), 'Exponental Multiplicative Schedule' )

schedule = LogarithmicMultiplicativeSchedule( T_start=t_start , T_freeze=t_freeze , alpha=6.00)
plot_schedule( schedule.get_schedule(N_steps), 'Logarithmic Multiplicative Schedule' )

schedule = LinearMultiplicativeSchedule( T_start=t_start , T_freeze=t_freeze , alpha=2.00)
plot_schedule( schedule.get_schedule(N_steps), 'Linear Multiplicative Schedule' )

schedule = LinearMultiplicativeSchedule( T_start=t_start , T_freeze=t_freeze , alpha=1.00)
plot_schedule( schedule.get_schedule(N_steps), 'Linear Multiplicative Schedule' )


schedule = QuadraticMultiplicativeSchedule( T_start=t_start , T_freeze=t_freeze , alpha=0.4)
plot_schedule( schedule.get_schedule(N_steps), 'Quadratic Multiplicative Schedule' )
```

### Additive Schedules

```python
t_start = 2.0
t_freeze = 0.15
dwell = 15000
N_steps = 120000

schedule = LinearAdditiveSchedule( T_start = t_start , T_freeze = t_freeze, dwell = dwell)
plot_schedule( schedule.get_schedule(N_steps), 'Linear Additive Schedule' )

schedule = QuadraticAdditiveSchedule( T_start = t_start, T_freeze = t_freeze, dwell = dwell)
plot_schedule( schedule.get_schedule(N_steps), 'Quadratic Additive Schedule' )

schedule = TrignometricAdditiveSchedule( T_start = t_start, T_freeze = t_freeze, dwell = dwell)
plot_schedule( schedule.get_schedule(N_steps), 'Trignometric Additive Schedule' )
```
